```python
import pandas as pd
#Ali'nin ders ve not bilgileri

data = {'Ders': ['Algoritma', 'BilgisayarProgramlama', 'Elektronik-2', 'AviyonikSistemler'],
        'Not': [95, 100, 65, 90]}
#Veriyi bir DataFrame'e dönüştür
df = pd.DataFrame(data)
#En yüksek notu bul
en_yuksek_not = df['Not'].max()
print("En yüksek not:", en_yuksek_not)
#Ortalama notu hesapla
ortalama_not = df['Not'].mean()
print("Ortalama not:", ortalama_not)
print(df.describe())
```

```python
import numpy as np

#Rastgele tanımlanan degerler-sıralaması olmayan
dizi = np.array([1, -55, 9, -23, 13, -21, 23, 37, 29, -6])

#Dizinin ortalama sonuc
ortalama = np.mean(dizi)
print("Ortalama:", ortalama)

#Dizinin standard sapma
std_sapma = np.std(dizi)
print("Standart Sapma:", std_sapma)

#Dizinin mutlağını al
mutlak_veriler = np.abs(dizi)
print("Mutlak Deger Cıktısı:", mutlak_veriler)
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)
sizes = 1000 * np.random.rand(50)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='viridis')
plt.title('Dağılım Grafiği')
plt.xlabel('X Ekseni')
plt.ylabel('Y Ekseni')
plt.colorbar()
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.linspace(-5, 5, 100)  # -5 ile 5 arasında eşit aralıklı 100 veri noktası oluşturur.
y = np.linspace(-5, 10, 50)  #-5 ile 10 arasında eşit aralıklı 50 veri noktası oluşturur.
X, Y = np.meshgrid(x, y)  #ızgara şeklini aktif ediyoruz.
Z = np.sin(np.sqrt(X**2 + Y**2))  # X ve Y üzerindeki her nokta için Z koordinatını hesaplar.
#Z değeri, x ve y'in karelerinin toplamının karekökünün sinüsüdür. Bu bir yüzey oluşturur.

ax.plot_surface(X, Y, Z, cmap='plasma')

ax.set_title('3D Yüzey Grafiği Örneği')
plt.show()
```

```python
#matpltlib'in tüm renkleri

import matplotlib.pyplot as plt
import numpy as np

#Renk haritası oluşturma fonksiyonu

def plot_colormaps():
    colormaps = plt.colormaps()
    n = len(colormaps)
    fig, axs = plt.subplots(n//2, 2, figsize=(10, 2*(n//2)))
    for ax, cmap_name in zip(axs.ravel(), colormaps):
        cmap = plt.get_cmap(cmap_name)
        img = np.linspace(0, 1, 256).reshape(1, -1)
        ax.imshow(img, aspect='auto', cmap=cmap)
        ax.set_title(cmap_name, fontsize=8)
        ax.set_axis_off()
    plt.show()

plot_colormaps()
```

**BeautifulSoup**

İnternet sayfasından başlıkları (h1, h2, h3) alır ve ekrana yazdırır.

```python
import requests
from bs4 import BeautifulSoup

url = 'https://milliyet.com.tr/'

response = requests.get(url)

if response.status_code == 200:
    soup = BeautifulSoup(response.content, 'html.parser')

    for heading in soup.find_all(['h1', 'h2', 'h3']):
        print(heading.text.strip())
else:
    print("Bağlantı başarısız: ", response.status_code)
```

```python
import requests
from bs4 import BeautifulSoup

url = 'https://alicetinkaya.site/'

response = requests.get(url)

if response.status_code == 200:
    soup = BeautifulSoup(response.content, 'html.parser')

    links = soup.find_all('a')

    for link in links:
        href = link.get('href')
        text = link.text.strip()
        if href:
            print("Bağlantı metni: ", text, "URL:", href)
    print("Toplam bulunan bilgi sayısı: ", len(links), " adet")
else:
    print("Bağlantı başarısız: ", response.status_code)
```
