```python
image = cv2.imread('Opencv_ornekleri/siberakademi_banner.png')
```

```python
#1.Aşama: Yuklenen görseli gösterme
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Yüklenen Görüntünün Gösterimi")
plt.show()
```

```python
#2.Aşama: Görüntüyü gri tonlamaya dönüştürme

gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
#Gri tonlanmıl görüntüyü gösterme
plt.imshow(gray_image, cmap='gray')
plt.title("Gri Tonlanmış Görüntü")
plt.show()
```

```python
#3.Aşama: Görüntüyü bulanıklaştırma
blurred_image = cv2.GaussianBlur(gray_image, (5, 5), 0)
#Bulanıklaştırılmış görüntüyü gösterme
plt.imshow(blurred_image, cmap='gray')
plt.title("Bulanıklaştırılmış Görüntü")
plt.show()
```

```python
# 4.aşama: Kenar Algılama (Canny Edege Detection)
edges = cv2.Canny(blurred_image, threshold1=100, threshold2=200)
#Kenar algılama sonucunu gösterme
plt.imshow(edges, cmap='gray')
plt.title("Kenar Algılama Sonucu")
plt.show()
```

```python
#5.Aşama: Tüm görüntülerin birlikte gösterilmesi
plt.figure(figsize=(10, 5))

plt.subplot(1, 4, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Orijinal Görüntü")

plt.subplot(1, 4, 2)
plt.imshow(gray_image, cmap='gray')
plt.title("Gri Tonlama")

plt.subplot(1, 4, 3)
plt.imshow(blurred_image, cmap='gray')
plt.title("Bulanık Görüntü")

plt.subplot(1, 4, 4)
plt.imshow(edges, cmap='gray')
plt.title("Kenar Algılama")

plt.show()
```

```python
#Görüntüleri yükleme
image1 = cv2.imread('Opencv_ornekleri/trafik-2.png')
image2 = cv2.imread('Opencv_ornekleri/trafik-3.png')
#Görüntüleri RGB formatına çevirme (OpenCV BGR formatında okur)
image1 = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)
image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)
#Görüntüleri gösterme
plt.subplot(1, 2, 1)
plt.imshow(image1)
plt.title('Görüntü 1')

plt.subplot(1, 2, 2)
plt.imshow(image2)
plt.title('Görüntü 2')

plt.show()
```

```python
#Görüntüleri aynı boyuta getirme (boyutlar farklıysa)
image1_resized = cv2.resize(image1, (image2.shape[1], image2.shape[0]))
#Görüntülri yan yana gösterme
plt.subplot(1, 2, 1)
plt.imshow(image1_resized)
height, width, channels = image1_resized.shape #1.Görüntünün boyutlarını bulma
plt.title('Yeniden Boyutlandırılan 1. Görsel')
print("1. Görüntü genişliği (x ekseni):", width, "piksel", "1. Görüntü yüksekliği (y ekseni):", height, "piksel")

plt.subplot(1, 2, 2)
plt.imshow(image2)
height2, width2, channels2 = image2.shape  #2.Görüntünün boyutlarını bulma
plt.title('Yeniden Boyutlandırılan 2. Görsel')
print("2. Görüntü genişliği (x ekseni):", width2, "piksel", "2. Görüntü yüksekliği (y ekseni):", height2, "piksel")

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image1 = cv2.imread('Opencv_ornekleri/trafik-2.png')
image2 = cv2.imread('Opencv_ornekleri/trafik-3.png')

hsv_image1 = cv2.cvtColor(image1, cv2.COLOR_BGR2HSV)
hsv_image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2HSV)

lower_red = np.array([0, 120, 70])
upper_red = np.array([10, 255, 255])

mask1_image1 = cv2.inRange(hsv_image1, lower_red, upper_red)
mask1_image2 = cv2.inRange(hsv_image2, lower_red, upper_red)

lower_red2 = np.array([170, 120, 70])
upper_red2 = np.array([180, 255, 255])

mask2_image1 = cv2.inRange(hsv_image1, lower_red2, upper_red2)
mask2_image2 = cv2.inRange(hsv_image2, lower_red2, upper_red2)

red_mask_image1 = mask1_image1 + mask2_image1
red_mask_image2 = mask1_image2 + mask2_image2

red_result_image1 = cv2.bitwise_and(image1, image1, mask=red_mask_image1)
red_result_image2 = cv2.bitwise_and(image2, image2, mask=red_mask_image2)

plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(red_result_image1, cv2.COLOR_BGR2RGB))
plt.title('Kırmızı Ok - Görüntü 1')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(red_result_image2, cv2.COLOR_BGR2RGB))
plt.title('Kırmızı Ok - Görüntü 2')

plt.show()
```
